```python
selected_columns = df[['How long have you been associated with or aware of the college?',
                       'How you rate the college overall?',
                       'How would you rate the academic standards of the college?',
                       'Are you satisfied with the teaching quality of the faculty?',
                       'What do you think about the curriculum and its relevance to industry standards?',
                       'Do you think the college provides adequate resources (library, labs, online resources, etc.) to support learning?',
                       'How would you rate the college’s infrastructure (classrooms, labs, internet facilities, etc.)?',
                       'Are extracurricular activities and sports adequately promoted at the college?',
                       'How satisfied are you with the cleanliness and maintenance of the campus?',
                       'How would you rate the career guidance and placement opportunities provided by the college?',
                       'Does the college provide adequate emotional and academic support to students?',
                       'Do you feel valued and respected as a student/community member of the college?',
                       'How likely are you to recommend this college to others?'

                       ]]
new_df = selected_columns.rename(columns = {'How long have you been associated with or aware of the college?':'College Association',
                                            'How you rate the college overall?':'College_rate',
                                            'How would you rate the academic standards of the college?':'Academic_standard',
                                            'Are you satisfied with the teaching quality of the faculty?':'Teaching_quality',
                                            'What do you think about the curriculum and its relevance to industry standards?':'Curriculum_relevance',
                                            'Do you think the college provides adequate resources (library, labs, online resources, etc.) to support learning?':'Academic_Resources',
                                            'How would you rate the college’s infrastructure (classrooms, labs, internet facilities, etc.)?':'College-infracture',
                                            'Are extracurricular activities and sports adequately promoted at the college?':'extracurricular-activities',
                                            'How satisfied are you with the cleanliness and maintenance of the campus?':'Hygene',
                                            'How would you rate the career guidance and placement opportunities provided by the college?':'Career_guide',
                                            'Does the college provide adequate emotional and academic support to students?':'Support',
                                            'Do you feel valued and respected as a student/community member of the college?':'Social_value',
                                            'How likely are you to recommend this college to others?':'Recommendation-others'
                                            })

new_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# One-hot encode nominal categorical columns
nominal_cols = ['College Association', 'Academic_Resources', 'extracurricular-activities', 'Support', 'Social_value']
df_encoded = pd.get_dummies(new_df, columns=nominal_cols, drop_first=True)
df_encoded.head(10)
```

```python
# Ordinal encode ordinal categorical columns
ordinal_cols = {
    'Academic_standard': ['Poor', 'Fair', 'Good', 'Excellent'],
    'Teaching_quality': ['Neutral', 'Satisfied', 'Dissatisfied', 'Very Satisfied', 'Very Dissatisfied'],
    'Curriculum_relevance': ['Not Sure', 'Irrevalent', 'Relavent', 'Very Relevant', 'Neutral'],
    'Career_guide': ['Poor', 'Fair', 'Good', 'Excellent'],
    'Recommendation-others': ['Very Unlikely', 'Unlikely', 'Neutral', 'Likely', 'Very likely']
}

# Handle NaN values before encoding
# Option 1: Impute NaN with the most frequent value
for col in ordinal_cols:
    # Calculate the most frequent value
    most_frequent_value = new_df[col].mode()[0]
    # Fill NaN with the most frequent value
    new_df[col].fillna(most_frequent_value, inplace=True)

# Option 2: Remove rows with NaN values (use with caution, might lead to data loss)
# for col in ordinal_cols:
#     new_df.dropna(subset=[col], inplace=True)

for col, categories in ordinal_cols.items():
    ordinal_encoder = OrdinalEncoder(categories=[categories])
    df_encoded[col] = ordinal_encoder.fit_transform(new_df[[col]])

# Display encoded data
print("\nEncoded DataFrame:")
print(df_encoded)
```

**Data CLeaning**

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in Open_ended_df.index:
  review = re.sub('[^a-zA-Z]', ' ', Open_ended_df.loc[i, 'Suggestion']) # Use .loc to access rows by index label
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

**Data Transformation**

**Creating a Bag of Words model**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1200)
```

```python
X = cv.fit_transform(corpus).toarray()    #feeding the corpus to model
y = df_encoded.loc[Open_ended_df.index, 'Recommendation-others'].values
#y = df.iloc[:, -1].values
```

```python
X.shape, y.shape
```

```python
#Saving BoW dictionary to later use in prediction
import pickle
bow_path = 'c1_BoW_Sentiment_Model.pkl'
pickle.dump(cv, open(bow_path, "wb"))
```

**Dividing datasets to train and test sets**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

**Min-Max Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Model fitting(Naive Based)**

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
print(classifier.fit(X_train, y_train))
# Exporting NB Classifier to later use in prediction
import joblib
joblib.dump(classifier, 'c2_Classifier_Sentiment_Model')
```

**Model Performance**

```python
# checking the accuracies
print("Training Accuracy :", classifier.score(X_train, y_train))
print("Testing Accuracy :", classifier.score(X_test, y_test))
```

```python
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy_score(y_test, y_pred)
```

```python
#Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

model = RandomForestClassifier()

# fitting the training data to the model
model.fit(X_train, y_train)

# predicting the test set results
y_pred = model.predict(X_test)

# checking the accuracies
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))

# making the confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy_score(y_test, y_pred)
```

```python
#Using NLP
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix

# creating the model
model = MLPClassifier()

# fitting the training data to the model
model.fit(X_train, y_train)

# predicting the test set results
y_pred = model.predict(X_test)

# checking the accuracies
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))

# making the confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

**VADAR**

We will use NLTK's SentimentIntensityAnalyzer to get the neg/neu/pos scores of the text.

* This uses the "Bag of Words" approcah
* stop words are removed
* each word is scored and combined to total score

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sia = SentimentIntensityAnalyzer()
```

```python
for item in tqdm(Open_ended_df['Strength'].values[:10]):
  print(item)
```

```python
sia.polarity_scores(Open_ended_df['Strength'][9])
```

```python
#The polarity score on the entire dataset
res={}
for i, item in tqdm(Open_ended_df.iterrows(), total=len(Open_ended_df)):
  text = item['Strength']
  #myid = item['id']
  res[i] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(res).T
vaders.head(10)
```

```python
# Convert res to DataFrame
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={'index':'Open_ended_df_index'})
#the merge function will now merge using the 'Open_ended_df_index'
vaders = vaders.merge(Open_ended_df, how='left', left_on='Open_ended_df_index', right_index=True)
vaders.tail()
```

```python
#plot VADERs result
ax = sns.barplot(data=vaders, x='Open_ended_df_index', y='compound')
ax.set_title('Compound Score')
plt.show()
```

```python
figs, axs = plt.subplots(1, 3, figsize=(12, 3))

# Create a new data frame from the existing vaders
# which has the mean of each category grouped by the index
df_pos = vaders.groupby('Open_ended_df_index')['pos'].mean().reset_index()
df_neu = vaders.groupby('Open_ended_df_index')['neu'].mean().reset_index()
df_neg = vaders.groupby('Open_ended_df_index')['neg'].mean().reset_index()

# Plot the positive sentiment
sns.barplot(data=df_pos, x='Open_ended_df_index', y='pos', ax=axs[0]) # x now uses the Open_ended_df_index and y now uses the pos data in the new dataframe.
axs[0].set_title('Positive')

# Plot the neutral sentiment
sns.barplot(data=df_neu, x='Open_ended_df_index', y='neu', ax=axs[1]) # x now uses the Open_ended_df_index and y now uses the neu data in the new dataframe.
axs[1].set_title('Neutral')
# Plot the negative sentiment
sns.barplot(data=df_neg, x='Open_ended_df_index', y='neg', ax=axs[2]) # x now uses the Open_ended_df_index and y now uses the neg data in the new dataframe.
axs[2].set_title('Negative')

#plt.tight_layout()
plt.show()
```
