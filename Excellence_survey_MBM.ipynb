```python
# Download required NLTK resources
nltk.download('stopwords')
all_stopwords = set(stopwords.words('english'))

# Function to clean text
def clean_responses(text):
    if isinstance(text, str):
        text = re.sub('[^a-zA-Z]', ' ', text)
        text = text.lower()
        text = text.split()
        text = [word for word in text if not word in all_stopwords]
        text = ' '.join(text)
        return text
    else:
        return ""  # Return empty string for non-string inputs
```

##**Fine Tune Bert**

```python
# Load model
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3)

# Set training arguments with improvements
training_args = TrainingArguments(
    output_dir="./results",          # Output directory for results
    #evaluation_strategy="epoch",     # Evaluate once per epoch
    #save_strategy="epoch",          # Save model at the end of each epoch to match evaluation strategy
    learning_rate=2e-5,              # Learning rate
    per_device_train_batch_size=16,  # Batch size for training
    per_device_eval_batch_size=16,   # Batch size for evaluation
    num_train_epochs=10,              # Increase number of epochs
    weight_decay=0.01,               # Strength of weight decay
    logging_dir="./logs",            # Directory for logging
    logging_steps=10,                # Log every 10 steps
    lr_scheduler_type="linear",      # Use linear learning rate scheduler with warmup
    warmup_steps=500,                # Number of warmup steps for learning rate scheduler
    #load_best_model_at_end=True,     # Load the best model at the end of training
    #metric_for_best_model="eval_loss", # Monitor eval loss to determine the best model
    save_total_limit=3,              # Limit the number of checkpoints to save
    gradient_accumulation_steps=1    # Simulate larger batch size if GPU memory is limited
)
# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)

# Fine-tuning the model
trainer.train()
```

```python
# Detection System (Example)
def detect_Sentiment(text):
    cleaned_text = clean_responses(text)
    inputs = tokenizer(cleaned_text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()
    return label_encoder.inverse_transform([predicted_class])[0]

# Test detection system with a mixture of normal, positive, or negative.
sample_texts = [
    "I feel perfectly fine with the college, nothing to worry about.",
    "I can't go wrong with learning.",
    "Needs lots of changes",
    "Not good teaching padageogy",
    "Canteen is awesome",
    "Need of frequent visit in company for better placement",
    "Need more works shops",
    "I hope there is a prayer in the merning with nationalantom",
    "The department is very helpful"
]

# Assuming the detect_anxiety function is already defined and takes a sentence as input
for text in sample_texts:
    predicted_class = detect_Sentiment(text)
    print(f"Sentence: {text}\nPredicted class: {predicted_class}\n")
```

```python
# Create topic mapping
topic_mapping = {v['topic_id']: k for k, v in df_config.items()}

# Common sentiment column name
sentiment_col = 'Sentiment, Positive = 1, Negative = -1, Neutral =0'
# Load and preprocess all dataframes
dataframes = {}
combined_df = None

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
import torch
from torch.utils.data import DataLoader, Dataset as TorchDataset
from transformers import BertTokenizer, Trainer, TrainingArguments, BertForSequenceClassification
from transformers import EarlyStoppingCallback
import re
import nltk
from nltk.corpus import stopwords
from imblearn.over_sampling import RandomOverSampler
from tqdm import tqdm

# Download required NLTK resources
nltk.download('stopwords', quiet=True)
all_stopwords = set(stopwords.words('english'))

# Function to clean text
def clean_responses(text):
    if isinstance(text, str):
        text = re.sub('[^a-zA-Z]', ' ', text)
        text = text.lower()
        text = text.split()
        text = [word for word in text if not word in all_stopwords]
        text = ' '.join(text)
        return text
    else:
        return ""  # Return empty string for non-string inputs

# Custom Dataset class
class SentimentDataset(TorchDataset):
    def __init__(self, encodings, labels, topics=None):
        self.encodings = encodings
        self.labels = labels
        self.topics = topics  # Add topics for tracking data source

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        if self.topics is not None:
            item['topic'] = self.topics[idx]
        return item

    def __len__(self):
        return len(self.labels)

# Set random seed for reproducibility
def set_seed(seed_value=42):
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed_value)

set_seed()

# Function to predict sentiment and topic for new text
def predict_sentiment_and_topic(text, model, tokenizer, topic_mapping):
    cleaned_text = clean_responses(text)
    inputs = tokenizer(cleaned_text, return_tensors="pt", truncation=True, padding=True, max_length=128)

    # Move to GPU if available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        sentiment_prediction = torch.argmax(logits[:, :3], dim=1).item()  # First 3 for sentiment
        topic_prediction = torch.argmax(logits[:, 3:], dim=1).item()      # Rest for topic

    # Map back to original sentiment and topic labels
    sentiment_map = {0: "Negative", 1: "Neutral", 2: "Positive"}
    predicted_sentiment = sentiment_map[sentiment_prediction]
    predicted_topic = topic_mapping[topic_prediction]

    # Get confidence scores
    sentiment_probs = torch.softmax(logits[0, :3], dim=0)
    topic_probs = torch.softmax(logits[0, 3:], dim=0)

    return {
        'sentiment': predicted_sentiment,
        'sentiment_confidence': sentiment_probs[sentiment_prediction].item(),
        'topic': predicted_topic,
        'topic_confidence': topic_probs[topic_prediction].item()
    }

try:
    # Load the data
    excel_file = 'Open_ended_responses.xlsx'

    # Define text and sentiment columns for each dataframe
    df_config = {
        'Suggestions': {'text_col': 'Suggestion', 'topic_id': 0},
        'Improvement': {'text_col': 'Improvement', 'topic_id': 1},
        'Additional_Comments': {'text_col': 'Comments', 'topic_id': 2},
        'Strength': {'text_col': 'Strength', 'topic_id': 3}
    }

    # Create topic mapping
    topic_mapping = {v['topic_id']: k for k, v in df_config.items()}

    # Common sentiment column name
    sentiment_col = 'Sentiment, Positive = 1, Negative = -1, Neutral =0'

    # Load and preprocess all dataframes
    dataframes = {}
    combined_df = None

    for sheet_name, config in df_config.items():
        print(f"Processing {sheet_name} sheet...")

        try:
            df = pd.read_excel(excel_file, sheet_name=sheet_name)

            # Verify columns exist
            text_col = config['text_col']
            if text_col not in df.columns:
                print(f"Warning: Column '{text_col}' not found in {sheet_name}. Checking for alternatives...")
                # Try to find a suitable text column
                text_columns = [col for col in df.columns if any(keyword in col.lower() for keyword in
                               ['text', 'comment', 'suggestion', 'improvement', 'strength', sheet_name.lower()])]
                if text_columns:
                    text_col = text_columns[0]
                    print(f"Using '{text_col}' as text column for {sheet_name}")
                    config['text_col'] = text_col
                else:
                    print(f"No suitable text column found in {sheet_name}. Skipping.")
                    continue

            if sentiment_col not in df.columns:
                print(f"Warning: Column '{sentiment_col}' not found in {sheet_name}. Checking for alternatives...")
                # Try to find a suitable sentiment column
                sentiment_columns = [col for col in df.columns if any(keyword in col.lower() for keyword in
                                    ['sentiment', 'polarity', 'emotion', 'feeling'])]
                if sentiment_columns:
                    sentiment_col = sentiment_columns[0]
                    print(f"Using '{sentiment_col}' as sentiment column for {sheet_name}")
                else:
                    print(f"No sentiment column found in {sheet_name}. Skipping.")
                    continue

            # Clean text data
            df[text_col] = df[text_col].apply(clean_responses)

            # Handle missing values
            df = df.dropna(subset=[text_col, sentiment_col])

            # Convert sentiment to integers
            df[sentiment_col] = pd.to_numeric(df[sentiment_col], errors='coerce')
            df = df.dropna(subset=[sentiment_col])
            df[sentiment_col] = df[sentiment_col].astype(int)

            # Add topic identifier
            df['topic_id'] = config['topic_id']
            df['topic_name'] = sheet_name

            # Add to dictionary of dataframes
            dataframes[sheet_name] = df

            # Add to combined dataframe
            if combined_df is None:
                combined_df = df[[text_col, sentiment_col, 'topic_id', 'topic_name']].copy()
            else:
                combined_df = pd.concat([
                    combined_df,
                    df[[text_col, sentiment_col, 'topic_id', 'topic_name']].copy()
                ])

            print(f"Added {len(df)} rows from {sheet_name}")

        except Exception as e:
            print(f"Error processing {sheet_name}: {str(e)}")

    if combined_df is None or len(combined_df) == 0:
        raise ValueError("No valid data found in any sheet")

    # Rename columns to standardize
    combined_df = combined_df.rename(columns={
        combined_df.columns[0]: 'text',
        sentiment_col: 'sentiment'
    })

    # Map sentiment values to 0, 1, 2 for BERT classification
    # -1=Negative (0), 0=Neutral (1), 1=Positive (2)
    label_mapping = {-1: 0, 0: 1, 1: 2}
    combined_df['sentiment_mapped'] = combined_df['sentiment'].map(label_mapping)

    # Collect metrics for each topic
    topic_metrics = {}
    all_test_texts = []
    all_test_sentiment_labels = []
    all_test_topic_labels = []
    all_sentiment_predictions = []
    all_topic_predictions = []

    # For individual topic training/evaluation
    for topic, df in dataframes.items():
        print(f"\nProcessing topic: {topic}")
        config = df_config[topic]
        text_col = config['text_col']

        # Map sentiment values
        df['sentiment_mapped'] = df[sentiment_col].map(label_mapping)

        # Balance the dataset
        X = df[[text_col]]
        y = df['sentiment_mapped']

        print(f"Class distribution before balancing for {topic}:")
        print(y.value_counts())

        # Apply random oversampling
        ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
        X_resampled, y_resampled = ros.fit_resample(X, y)

        # Create balanced dataframe
        balanced_df = X_resampled.copy()
        balanced_df['sentiment_mapped'] = y_resampled

        print(f"Class distribution after balancing for {topic}:")
        print(balanced_df['sentiment_mapped'].value_counts())

        # Split the data
        train_text, test_text, train_label, test_label = train_test_split(
            balanced_df[text_col],
            balanced_df['sentiment_mapped'],
            test_size=0.2,
            random_state=42,
            stratify=balanced_df['sentiment_mapped']
        )

        # Save test data for later comparison
        test_topic_labels = [config['topic_id']] * len(test_text)

        all_test_texts.extend(test_text.tolist())
        all_test_sentiment_labels.extend(test_label.tolist())
        all_test_topic_labels.extend(test_topic_labels)

        topic_metrics[topic] = {
            'test_texts': test_text.tolist(),
            'test_labels': test_label.tolist(),
            'test_topic_labels': test_topic_labels
        }

    print("\n--- Creating combined model ---")

    # Create datasets for combined model (sentiment + topic classification)
    # For the combined model, we'll use a multi-task approach:
    # - First 3 outputs for sentiment (neg, neu, pos)
    # - Next 4 outputs for topic classification

    # Balance the combined dataset
    X_combined = combined_df[['text']]
    y_combined = pd.DataFrame({
        'sentiment': combined_df['sentiment_mapped'],
        'topic': combined_df['topic_id']
    })

    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(
        X_combined, y_combined,
        test_size=0.2,
        random_state=42,
        stratify=y_combined['topic']  # Stratify by topic first
    )

    # Debug what's being passed to tokenizer
    print(f"Type of data being tokenized: {type(X_train['text'])}")
    print(f"First few examples: {X_train['text'].head().tolist()}")

    # Load tokenizer
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    # Tokenize the data
    # Before tokenizing, make sure the text column is properly extracted
    print(f"X_train columns: {X_train.columns}")  # Debug to see what columns are available

    # Check if the text column exists
    if 'text' not in X_train.columns:
        # If there's only one column, it's likely our text column
        if len(X_train.columns) == 1:
            text_column = X_train.columns[0]
            print(f"Using '{text_column}' as the text column")
            train_texts = X_train[text_column].fillna("").astype(str).tolist()
        else:
            raise ValueError("No 'text' column found in X_train and can't determine which column to use")
    else:
        train_texts = X_train['text'].fillna("").astype(str).tolist()

    # Make sure we have valid strings to tokenize
    train_texts = [text if isinstance(text, str) else "" for text in train_texts]

    # Now tokenize with validated text data
    train_encodings = tokenizer(
        train_texts,
        truncation=True,
        padding=True,
        max_length=128,
        return_tensors='pt'
        )

    if 'text' not in X_test.columns:
      if len(X_test.columns) == 1:
        text_column = X_test.columns[0]
        test_texts = X_test[text_column].fillna("").astype(str).tolist()
      else:
        raise ValueError("No 'text' column found in X_test and can't determine which column to use")
    else:
      test_texts = X_test['text'].fillna("").astype(str).tolist()

    test_texts = [text if isinstance(text, str) else "" for text in test_texts]

    test_encodings = tokenizer(
        test_texts,
        truncation=True,
        padding=True,
        max_length=128,
        return_tensors='pt'
        )

    # Custom model class for multi-task learning (sentiment + topic)
    class BertForMultiTask(BertForSequenceClassification):
        def __init__(self, config):
            super().__init__(config)
            # We'll have 7 outputs: 3 for sentiment, 4 for topic
            self.num_labels = 7
            self.classifier = torch.nn.Linear(config.hidden_size, 7)

        def forward(self, input_ids=None, attention_mask=None, labels=None, **kwargs):
            outputs = self.bert(
                input_ids=input_ids,
                attention_mask=attention_mask,
                **kwargs
            )

            pooled_output = outputs.pooler_output
            logits = self.classifier(pooled_output)

            loss = None
            if labels is not None:
                loss_fct = torch.nn.CrossEntropyLoss()
                sentiment_labels = labels[:, 0]
                topic_labels = labels[:, 1]

                # Loss for sentiment (first 3 outputs)
                sentiment_loss = loss_fct(logits[:, :3], sentiment_labels)

                # Loss for topic (next 4 outputs)
                topic_loss = loss_fct(logits[:, 3:], topic_labels)

                # Combined loss (weighted sum)
                loss = sentiment_loss + topic_loss

            return torch.nn.Linear.Output(
                loss=loss,
                logits=logits,
                hidden_states=outputs.hidden_states,
                attentions=outputs.attentions
            )

    # Create a standard BERT model (since custom class has challenges)
    # We'll use a workaround for multi-task learning
    num_sentiment_labels = 3
    num_topic_labels = 4
    model = BertForSequenceClassification.from_pretrained(
        'bert-base-uncased',
        num_labels=num_sentiment_labels + num_topic_labels  # 7 total outputs
    )

    # Prepare datasets with both sentiment and topic labels
    class MultiTaskDataset(TorchDataset):
        def __init__(self, encodings, sentiment_labels, topic_labels):
            self.encodings = encodings
            self.sentiment_labels = sentiment_labels
            self.topic_labels = topic_labels

        def __getitem__(self, idx):
            item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
            # We'll use the 'labels' field for sentiment
            item['labels'] = torch.tensor(self.sentiment_labels[idx])
            # And add a custom field for topic
            item['topic_labels'] = torch.tensor(self.topic_labels[idx])
            return item

        def __len__(self):
            return len(self.sentiment_labels)

    # Create datasets
    train_dataset = MultiTaskDataset(
        {k: v.numpy() for k, v in train_encodings.items()},
        y_train['sentiment'].tolist(),
        y_train['topic'].tolist()
    )

    test_dataset = MultiTaskDataset(
        {k: v.numpy() for k, v in test_encodings.items()},
        y_test['sentiment'].tolist(),
        y_test['topic'].tolist()
    )

    # Custom trainer for multi-task learning
    class MultiTaskTrainer(Trainer):
        def compute_loss(self, model, inputs, return_outputs=False):
            labels = inputs.pop("labels")
            topic_labels = inputs.pop("topic_labels")

            outputs = model(**inputs)
            logits = outputs.logits

            # Split logits into sentiment and topic parts
            sentiment_logits = logits[:, :num_sentiment_labels]
            topic_logits = logits[:, num_sentiment_labels:]

            # Compute losses
            loss_fct = torch.nn.CrossEntropyLoss()
            sentiment_loss = loss_fct(sentiment_logits.view(-1, num_sentiment_labels), labels.view(-1))
            topic_loss = loss_fct(topic_logits.view(-1, num_topic_labels), topic_labels.view(-1))

            # Combined loss - you can weight these differently if needed
            loss = sentiment_loss + topic_loss

            return (loss, outputs) if return_outputs else loss

        def prediction_step(self, model, inputs, prediction_loss_only=False):
            labels = inputs.get("labels")
            topic_labels = inputs.get("topic_labels")

            # Remove topic_labels from inputs for model forward pass
            inputs_copy = {k: v for k, v in inputs.items() if k != "topic_labels"}

            with torch.no_grad():
                outputs = model(**inputs_copy)
                logits = outputs.logits

                # For evaluation, we'll return both sets of predictions
                sentiment_preds = logits[:, :num_sentiment_labels].argmax(dim=-1)
                topic_preds = logits[:, num_sentiment_labels:].argmax(dim=-1)

                # Combine predictions for metrics calculation
                combined_preds = torch.stack([sentiment_preds, topic_preds], dim=1)
                combined_labels = torch.stack([labels, topic_labels], dim=1) if labels is not None else None

                loss = None
                if labels is not None:
                    # Call the compute_loss method to get loss value
                    inputs_with_labels = {**inputs_copy, "labels": labels, "topic_labels": topic_labels}
                    loss = self.compute_loss(model, inputs_with_labels)

            return (loss, combined_preds, combined_labels)

    # Define metrics for evaluation
    def compute_metrics(pred):
        combined_preds = pred.predictions
        combined_labels = pred.label_ids

        # Extract sentiment predictions and labels
        sentiment_preds = combined_preds[:, 0]
        sentiment_labels = combined_labels[:, 0]

        # Extract topic predictions and labels
        topic_preds = combined_preds[:, 1]
        topic_labels = combined_labels[:, 1]

        # Calculate sentiment metrics
        sentiment_acc = accuracy_score(sentiment_labels, sentiment_preds)
        sentiment_f1 = f1_score(sentiment_labels, sentiment_preds, average='macro')

        # Calculate topic metrics
        topic_acc = accuracy_score(topic_labels, topic_preds)
        topic_f1 = f1_score(topic_labels, topic_preds, average='macro')

        # Combined metrics
        overall_acc = (sentiment_acc + topic_acc) / 2

        return {
            'sentiment_accuracy': sentiment_acc,
            'sentiment_f1': sentiment_f1,
            'topic_accuracy': topic_acc,
            'topic_f1': topic_f1,
            'overall_accuracy': overall_acc
        }

    # Set up training arguments
    training_args = TrainingArguments(
        output_dir='./results_multi_task',
        num_train_epochs=5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        learning_rate=3e-5,
        weight_decay=0.01,
        logging_dir='./logs_multi_task',
        logging_steps=10,
        #evaluation_strategy='epoch',
        #save_strategy='epoch',
        #load_best_model_at_end=True,
        #metric_for_best_model='overall_accuracy',
        save_total_limit=2,
        gradient_accumulation_steps=2,
        warmup_ratio=0.1,
        report_to='none',
    )

    # Initialize the trainer
    trainer = MultiTaskTrainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=test_dataset,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
    )

    # Train the model
    print("Training multi-task model...")
    trainer.train()

    # Evaluate the model
    print("Evaluating model...")
    eval_results = trainer.evaluate()
    print(f"Evaluation results: {eval_results}")

    # Make predictions on test data
    print("Making predictions...")
    test_pred = trainer.predict(test_dataset)
    combined_predictions = test_pred.predictions

    # Extract sentiment and topic predictions
    sentiment_predictions = combined_predictions[:, 0]
    topic_predictions = combined_predictions[:, 1]

    # Extract sentiment and topic labels
    sentiment_labels = y_test['sentiment'].values
    topic_labels = y_test['topic'].values

    # Map back to original sentiment values for interpretability
    reverse_mapping = {0: -1, 1: 0, 2: 1}
    predicted_sentiments = [reverse_mapping[pred] for pred in sentiment_predictions]
    actual_sentiments = [reverse_mapping[label] for label in sentiment_labels]

    # Overall metrics
    print("\n------ OVERALL METRICS ------")

    # Sentiment classification report
    print("\nSentiment Classification Report:")
    print(classification_report(sentiment_labels, sentiment_predictions,
                             target_names=['Negative', 'Neutral', 'Positive']))

    # Topic classification report
    print("\nTopic Classification Report:")
    print(classification_report(topic_labels, topic_predictions,
                             target_names=list(topic_mapping.values())))

    # Sentiment confusion matrix
    sentiment_cm = confusion_matrix(sentiment_labels, sentiment_predictions)

    # Topic confusion matrix
    topic_cm = confusion_matrix(topic_labels, topic_predictions)

    # Prepare per-topic metrics
    topic_results = {}
    for topic_id, topic_name in topic_mapping.items():
        # Get indices for this topic
        topic_indices = (topic_labels == topic_id)

        if sum(topic_indices) > 0:  # Check if there are any samples for this topic
            # Get sentiment metrics for this topic
            topic_sentiment_acc = accuracy_score(
                sentiment_labels[topic_indices],
                sentiment_predictions[topic_indices]
            )

            topic_sentiment_f1 = f1_score(
                sentiment_labels[topic_indices],
                sentiment_predictions[topic_indices],
                average='macro'
            )

            topic_results[topic_name] = {
                'accuracy': topic_sentiment_acc,
                'f1_score': topic_sentiment_f1,
                'sample_count': sum(topic_indices)
            }

    # Create comparison plot
    plt.figure(figsize=(15, 10))

    # Accuracy comparison
    plt.subplot(2, 2, 1)
    topic_names = list(topic_results.keys())
    accuracies = [topic_results[t]['accuracy'] for t in topic_names]

    bars = plt.bar(topic_names, accuracies)
    plt.title('Sentiment Accuracy by Topic')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                 f'{height:.2f}', ha='center', va='bottom')

    # F1 Score comparison
    plt.subplot(2, 2, 2)
    f1_scores = [topic_results[t]['f1_score'] for t in topic_names]

    bars = plt.bar(topic_names, f1_scores)
    plt.title('Sentiment F1 Score by Topic')
    plt.ylabel('F1 Score')
    plt.ylim(0, 1)
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                 f'{height:.2f}', ha='center', va='bottom')

    # Sentiment confusion matrix
    plt.subplot(2, 2, 3)
    sns.heatmap(sentiment_cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Negative', 'Neutral', 'Positive'],
                yticklabels=['Negative', 'Neutral', 'Positive'])
    plt.title('Overall Sentiment Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')

    # Topic confusion matrix
    plt.subplot(2, 2, 4)
    sns.heatmap(topic_cm, annot=True, fmt='d', cmap='Greens',
                xticklabels=list(topic_mapping.values()),
                yticklabels=list(topic_mapping.values()))
    plt.title('Topic Classification Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')

    plt.tight_layout()
    plt.savefig('sentiment_topic_analysis_results.png')
    plt.show()

    # Save the model
    model_path = './sentiment_topic_model'
    model.save_pretrained(model_path)
    tokenizer.save_pretrained(model_path)
    print(f"Model saved to {model_path}")

    # Create a function to predict sentiment and topic for new input text
    def test_prediction(text):
        result = predict_sentiment_and_topic(text, model, tokenizer, topic_mapping)
        print(f"\nInput text: '{text}'")
        print(f"Predicted sentiment: {result['sentiment']} (confidence: {result['sentiment_confidence']:.2f})")
        print(f"Predicted topic: {result['topic']} (confidence: {result['topic_confidence']:.2f})")
        return result

    # Demo with a few sample inputs
    print("\n------ DEMO PREDICTIONS ------")
    test_prediction("The product works really well and I'm happy with it")
    test_prediction("The service was terrible and I want a refund")
    test_prediction("I think there could be improvements to the user interface")
    test_prediction("Your team provides excellent customer service")

    print("\n------ MODEL READY FOR USE ------")
    print("You can now use the model to classify new text inputs by topic and sentiment.")
    print("Function: test_prediction(text) is available for testing.")

except Exception as e:
    import traceback
    print(f"An error occurred: {str(e)}")
    traceback.print_exc()
```
